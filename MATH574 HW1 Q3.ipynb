```python
def FindCoeffs (root, rad):
    invdF=np.linalg.inv(dF(root))
    #print ("A is :",invdF)
    Y0=np.linalg.norm(np.matmul(invdF, F(root)))
    #print ("So Y0 is: ",Y0)
    Z2=FindZ2(root, rad, invdF)
    #print ( "And Z2 is", Z2)
    return [Y0, Z2]
```
